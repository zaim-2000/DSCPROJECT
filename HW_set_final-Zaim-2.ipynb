***Question #9 (6 pts)*** Zaim
- Write a function mean(numbers) that returns the average.
- Create an object--a list of numbers.
- Call the function using the object and print the result.

```python
def mean(numbers): # Creates function that sums the numbers in a list/tuple and divides them by number of numbers.
    
    """ 
    This sums the numbers in a list or tuple and then divides them 
    by the number of elements in the list aka taking the mean value.
    """
    answer= sum(numbers)/len(numbers)
    return answer
    
z= [1, 2, 9, 4] #creates object- a list of numbers
print(z)
print(f"Average of the list z above is {mean(z)}") # calls mean function on list z
```

Write a function that classifies a number:

- "low" if x < 0.3, "high" otherwise
- create an object or instance, a single number
- call the function and print the result

```python
def classify_number(x):
    """
    This function takes a number x and classifies it as either
    'low' if it is less than 0.3, or 'high' otherwise.
    """
    
    # Classifies  x as low if it's less than 0.3 and high otherwise
    if x < 0.3:
        return "low"   # Return "low" if condition is true
    else:
        return "high"  # Return "high" if condition is false


# Create a number (object/instance)
num = 0.25
print(num)

# Call the function on number =0.25 and print the result
print(classify_number(num))
```

- create a dictionary for the prices of banana, apple and organge (assign prices)
- create a new dictionary using comprehension with prices increased by 10%.

```python
prices = {"banana": 1.00,
          "apple": 1.50,
          "orange": 1.20} # Create a dictionary with original prices for banana, apple and orange
print(prices)
new_prices = {item: price * 1.10 for item, price in prices.items()} 
# Create a new dictionary with prices increased by 10%
print(new_prices)
```
